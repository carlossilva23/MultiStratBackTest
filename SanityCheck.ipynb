```python
import yfinance as yf

asset = yf.download("SPY", period="6mo")
clean_asset = asset["Close"].dropna()
clean_asset.rename(columns={"SPY": "Price"})
```

```python
beg = clean_asset.iloc[0]
begP = beg["Price"]
end = clean_asset.iloc[-1]
endP = end.iloc["Price"]

totalReturn = (endP - begP) / begP
totalReturn
```
